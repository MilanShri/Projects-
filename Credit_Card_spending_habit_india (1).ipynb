```python
# Ques 1) Total amount spend by Gender
```

```python
amount = df.groupby('Gender')['Amount'].sum()
```

```python
print("Total amount spend by Gender")
print(amount)
```

```python
#Ques2 ) Avg amount speed group by card type.

amount_on_card = df.groupby("Card Type")["Amount"].mean()
```

```python
print("Avg Amount speed through Card type")
print(amount_on_card)
```

```python
#ques 3) Total amount speed my city

amount_spend_by_city = df.groupby("City")["Amount"].sum().sort_values(ascending=False)
```

```python
amount_spend_by_city.head(5)
```

```python
amount_spend_by_city.tail(5)
```

```python
df
```

```python
#ques 4) amount spend based on exp type. 
amount_spend_on_exp_type = df.groupby("Exp Type")["Amount"].sum().sort_values(ascending=False)
```

```python
print("Total Amount Spend based on Exp type")
print(amount_spend_on_exp_type)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize= (12,6))

# total amount spend by gender
plt.subplot(2,2,1)
sns.barplot(x = amount.index , y = amount.values)
plt.title("total Amount spend based on Gender")
plt.xlabel("Gender")
plt.ylabel("Amount")

# Average transaction amount by card type
plt.subplot(2,2,2)
sns.barplot(x= amount_on_card.index, y= amount_on_card.values)
plt.title("Average transaction amount by card type")
plt.xlabel("card type")
plt.ylabel("Amount based on card")

#Total amount spent by city
plt.subplot(2,2,3)
sns.barplot(x = amount_spend_by_city.head(10).index ,y = amount_spend_by_city.head(10).values)
plt.title("Amount based on City")
plt.xlabel("City's")
plt.ylabel("Amount based on City")
plt.xticks(rotation = 90)

#Total amount spent by expense type
plt.subplot(2,2,4)
sns.barplot(x = amount_spend_on_exp_type.index ,y = amount_spend_on_exp_type.values)
plt.title("Amount Spend Based on Exp Type")
plt.xlabel("Exp Type")
plt.ylabel("Amount based on Ext type")

plt.tight_layout()
plt.show()
```
